```python
def printMatrix(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            print(matrix[i][j], end=' ')
        print()

def matrixAdd(A, B):
    n = len(A)
    m = len(A[0])
    C = [[None]*m for row in range(n)]# n x m
    for i in range(n):
        for j in range(m):
            C[i][j] = A[i][j]+B[i][j]
    return C

def matrixSubtract(A, B):
    n = len(A)
    m = len(A[0])
    C = [[None]*m for row in range(n)]# n x m
    for i in range(n):
        for j in range(m):
            C[i][j] = A[i][j]-B[i][j]
    return C

matrixA = [[1,3,5], [7,9,11], [13,15,17]]
matrixB = [[9,8,7], [6,5,4], [3,2,1]]
matrixC = matrixAdd(matrixA,matrixB)
matrixD = matrixSubtract(matrixA,matrixB)

print("MatrixA:")
printMatrix(matrixA)
print("MatrixB:")
printMatrix(matrixB)
print("MatrixA + MatrixB:")
printMatrix(matrixC)
print("MatrixA - MatrixB:")
printMatrix(matrixD)
```

```python
def printMatrix(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            print(matrix[i][j], end=' ')
        print()

def matrixMutiply(A, B):
    m = len(A)#a row  2
    na = len(A[0])#a colun 3
    nb = len(B)#b row  3
    p = len(B[0])#b colum 2
    if(na!=nb):
        print('ABçŸ©é™£ç„¡æ³•ç›¸ä¹˜')
        return
    C = [[None]*p for row in range(m)]
    for i in range(m):
        for j in range(p):
            tmp=0
            for k in range(na):
                tmp = tmp + A[i][k]*B[k][j]
            C[i][j] = tmp
    return C


matrixA = [[6,3,5], [8,9,7]]
matrixB = [[5,10], [14,7], [6,8]]
matrixC = matrixMutiply(matrixA,matrixB)

print("MatrixA:")
printMatrix(matrixA)
print("MatrixB:")
printMatrix(matrixB)
print("MatrixA * MatrixB:")
printMatrix(matrixC)
```

```python
def rotate(matrixA):
    matrixB = []
    r = len(matrixA)
    c = len(matrixA[0])
    for i in range(c):
        row = []
        for j in range(r-1, -1, -1):
            row.append(matrixA[j][i]) #columä¸è®Š rowç”±æœ€å¾Œä¸€ç­†é–‹å§‹å¾€ä¸Š
        matrixB.append(row)
    return matrixB

def printMatrix(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            print(matrix[i][j], end=' ')
        print()
#é‚„åŸ
def rotate2(matrixA):
    matrixB = []
    r = len(matrixA)       
    c = len(matrixA[0])
    for i in range(c-1,-1,-1):
        row = []
        for j in range(r):
            row.append(matrixA[j][i])
        matrixB.append(row)
    return matrixB

m1 = [[3,6],[2,5],[1,4]]
m2 = rotate(m1)
m3 = rotate2(m2)
printMatrix(m1)

print('æ—‹è½‰å¾Œ...')
printMatrix(m2)
print('===========')
printMatrix(m3)
```
